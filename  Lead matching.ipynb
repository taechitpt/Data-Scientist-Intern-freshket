### Data Preparation

```python
data['lower_Restaurant_name'] = data['Restaurant name'].apply(lambda x: x.lower().strip())
data.head()
```

```python
data['remove_special_text'] = data['lower_Restaurant_name'].apply(lambda x: re.sub(r'[^a-zA-Z0-9\u0E00-\u0E7F\s]',"",x))
data.head(10)
```

```python
stopword = ['ร้าน','สาขา','แถว','ซอย','บริษัท','เดอะ']
```

```python
data['Cleaned_Restaurant_name'] = data['remove_special_text'].apply(lambda x: [word for word in x.split() if word not in stopword])
data['Cleaned_Restaurant_name'] = data['Cleaned_Restaurant_name'].apply(lambda x: " ".join(x))
data.head(10)
```

-------------------------------

### Name Matching

- This section uses fuzzy matching to analyze the similarity between restaurant names. The similarity score is calculated using a fuzzy matching function, which returns a value between 0 and 100. A similarity threshold of 80 is applied to determine potential duplicate restaurant names. If the similarity score between two restaurant names is greater than or equal to 80, the pair is considered a potential match and append to name_matching.

```python
from rapidfuzz import fuzz
```

```python
name_matching = []
Cleaned_Restaurant_name = data['Cleaned_Restaurant_name']

for i in range(len(Cleaned_Restaurant_name)):
    for j in range(i+1,len(Cleaned_Restaurant_name)):
        name_1 = data.loc[i,'Cleaned_Restaurant_name']
        name_2 = data.loc[j,'Cleaned_Restaurant_name']
        score = fuzz.token_ratio(name_1,name_2)
        if score >= 80:
            name_matching.append((i,j,score))
name_matching
```

**- Precision/Recall trade-offs**

1. A higher threshold (e.g., 90) -> increases precision because only very similar names will be matched and reduced false positives, but can may miss true duplicates with spelling variations.
2. A lower threshold (e.g., 70) -> increases recall by capturing more potential duplicates. However, this may introduce false matches between restaurants with partially similar names.

**So thresholds in the range of 80–85 are often used in practice to balance precision and recall when performing fuzzy name matching.**

### Location Matching

```python
from geopy.distance import geodesic
```

- Location similarity is evaluated using a distance threshold of 200 meters. Restaurant pairs within 200 meters are considered geographically similar.

```python
location_matching = []
for i in range(len(data)):
    for j in range(i+1,len(data)):
        lat_1 = data.loc[i,'lat']
        long_1 = data.loc[i,'long']
        lat_2 = data.loc[j,'lat']
        long_2 = data.loc[j,'long']
        distance = geodesic((lat_1, long_1), (lat_2, long_2)).meters
        if distance <= 200:
            location_matching.append((i,j,distance))
location_matching
        
    
```

### Combined Matching & Evaluation

**name_similarlity_score**

```python
name_similarlity_score = []
Cleaned_Restaurant_name = data['Cleaned_Restaurant_name']

for i in range(len(Cleaned_Restaurant_name)):
    for j in range(i+1,len(Cleaned_Restaurant_name)):
        name_1 = data.loc[i,'Cleaned_Restaurant_name']
        name_2 = data.loc[j,'Cleaned_Restaurant_name']
        score = fuzz.token_ratio(name_1,name_2)
        name_similarlity_score.append((i,j,score))
name_similarlity_score 
```

----------------------------------

**location_similarity_score**

```python
location_similarity_score = []
for i in range(len(data)):
    for j in range(i+1,len(data)):
        lat_1 = data.loc[i,'lat']
        long_1 = data.loc[i,'long']
        lat_2 = data.loc[j,'lat']
        long_2 = data.loc[j,'long']
        distance = geodesic((lat_1, long_1), (lat_2, long_2)).meters
        score = max(0, 100 * (1 - distance/200))
        location_similarity_score.append((i,j,score))
location_similarity_score
        
```

----------------------------------------

- Combine score with solution :  ***final_score = (0.3 x name_similarlity_score) + (0.7 x location_similarity_score)***
  
In this approach, a higher weight is assigned to location similarity because restaurant names can sometimes be inconsistent due to spelling variations, abbreviations, or formatting differences. For example, the same restaurant may appear as “Sushi Hiro”, “Sushi Hero”, or “Sushi Hiro Thonglor”.

Location information, on the other hand, tends to be more stable and reliable for identifying the same physical restaurant. Therefore, location similarity is given a higher weight (0.7), while name similarity is given a lower weight (0.3).


```python
final_similarity_score = []

Cleaned_Restaurant_name = data['Cleaned_Restaurant_name']

for i in range(len(Cleaned_Restaurant_name)):
    for j in range(i+1,len(Cleaned_Restaurant_name)):
        name_1 = data.loc[i,'Cleaned_Restaurant_name']
        name_2 = data.loc[j,'Cleaned_Restaurant_name']
        name_score = fuzz.token_ratio(name_1,name_2)

        lat_1 = data.loc[i,'lat']
        long_1 = data.loc[i,'long']
        lat_2 = data.loc[j,'lat']
        long_2 = data.loc[j,'long']
        distance = geodesic((lat_1, long_1), (lat_2, long_2)).meters
        location_score = max(0, 100 * (1 - distance/200))

        final_score = (0.3*name_score)+(0.7*location_score)
        if final_score >=80:
            final_similarity_score.append((i,j,final_score))
final_similarity_score
        
```

### Describe evaluation metrics (precision, recall, F1-score)

- **Precision** measures the proportion of predicted duplicate restaurant pairs that are actually correct.
- **Recall** measures the proportion of true duplicate restaurants that the system successfully identifies.
- **F1-score** combines precision and recall into a single metric to evaluate the overall performance of the duplicate detection system.
